```python
class Stack:
    def __init__(self):
        self.items = []
        
    def push(self, element):
        self.items.append(element)
        
    def pop(self):
        return self.items.pop()
    
    def __len__(self):
        return len(self.items)
    
    def peek(self):
        return self.items[len(self.items) - 1]
    
    def is_empty(self):
        return self.items == []
```

```python
numbers = [1,2,3,4,5,6]

stack = Stack()

for number in numbers:
    stack.push(number)
    
out = []

while not stack.is_empty():
    out.append(stack.pop())
    
print(out)
```

```python
# Data cell
class Node:
    def __init__(self, data):
        self.data = data
        self.above = None
        
    def __str__(self):
        return str(self.data)
        
# Stack implementation
class Stack:
    def __init__(self):
        self.base = None # start point of stack
        self.__lenght = 0  # quantity of elements in the stack
        
    def push(self, element):
        new_node = Node(element) # create new node to add
        
        # in case we have an empty stack we define the new node as the base node, and increment the lenght
        if self.is_empty():
            self.base = new_node
            self.__lenght += 1
            return
        # otherwise we search for the last element and add the new node to it, and increment the length
        current = self.peek()
        current.above = new_node
        self.__lenght += 1
        
    def pop(self):
        # validate is stack isn't empty
        if self.is_empty():
            print('No elements to pop')
            return
        
        # in case we just have one node (base node) we reset the lenght and clear the base node
        if self.base.above is None:
            current_base = self.base
            self.base = None
            self.__lenght = 0
            return current_base
            
        # otherwise we loop through the stack to find the last node and the previous last node
        current = self.base
        last = self.base.above

        # until last node has a node we continue looping
        while last.above:
            current = last
            last = last.above

        # once we get the last, we remove the above to the previous and return the last node
        current.above = None
        self.__lenght -= 1
        return last
    
    def is_empty(self):
        # if we dont have a base node our stack is empty
        return self.base == None  
    
    def peek(self):
        # if our stack is empty we return none
        if self.is_empty():
            return None
        
        last = self.base
        # let's loop through the above of each last node until we dont have an above node
        # and then just return the node we got
        while last.above:
            last = last.above
        return last
        
    def __str__(self):
        # if we want convert our stack to str, we just return the last element
        return str(self.peek())
    
    def __len__(self):
        # in case we want know the lenght of the stack we just return the lenght property
        return self.__lenght
```

```python
stack = Stack()
print('Is stack empty? A: '+ str(stack.is_empty()))
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

print('Stack lenght: ' + str(len(stack)))
print('Last stack element: ' + str(stack.peek()))
print('Is stack empty? A: '+ str(stack.is_empty()))

print('Popped element: ' + str(stack.pop()))
print('Last stack element: ' + str(stack))
print('Stack lenght: ' + str(len(stack)))
```

```python
# Data cell
class Queue:
    def __init__(self):
        self.items = []
        
    def enqueue(self, element):
        self.items.append(element)
        
    def dequeue(self):
        return self.items.pop(0)
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        return self.items[len(self.items) - 1]
    
    def __len__(self):
        return len(self.items)
```

```python
queue = Queue()

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)

print('Queue lenght: ' + str(len(queue)))
print('Last queue element: ' + str(queue.peek()))

print('Is queue empty? A: '+ str(queue.is_empty()))

print('Popped element: ' + str(queue.dequeue()))
print('Last queue element: ' + str(queue.peek()))
print('Queue lenght: ' + str(len(queue)))
```

```python
# Extra challenge
# Data cell
class Node:
    def __init__(self, data):
        self.data = data
        self.above = None
        
    def __str__(self):
        return str(self.data)
        
# Stack implementation
class Queue:
    def __init__(self):
        self.base = None # start point of queue
        self.__lenght = 0  # quantity of elements in the queue
        
    def queue(self, element):
        new_node = Node(element) # create new node to add
        
        # in case we have an empty queue we define the new node as the base node, and increment the lenght
        if self.is_empty():
            self.base = new_node
            self.__lenght += 1
            return
        # otherwise we search for the last element and add the new node to it, and increment the length
        current = self.peek()
        current.above = new_node
        self.__lenght += 1
        
    def dequeue(self):
        # validate is queue isn't empty
        if self.is_empty():
            print('No elements to dequeue')
            return
        
        # in case we just have one node (base node) we reset the lenght and clear the base node
        if self.base.above is None:
            current_base = self.base
            self.base = None
            self.__lenght = 0
            return current_base
            
        # otherwise we save our first element to be able to return it
        # we set our above element as the base now
        # and we decrease the lenght
        out = self.base
        self.base = self.base.above
        self.__lenght -= 1
        return out
    
    def is_empty(self):
        # if we dont have a base node our queue is empty
        return self.base == None  
    
    def peek(self):
        # if our queue is empty we return none
        if self.is_empty():
            return None
        
        last = self.base
        # let's loop through the above of each last node until we dont have an above node
        # and then just return the node we got
        while last.above:
            last = last.above
        return last
        
    def __str__(self):
        # if we want convert our queue to str, we just return the first element
        return str(self.base)
    
    def __len__(self):
        # in case we want know the lenght of the queue we just return the lenght property
        return self.__lenght
```

```python
queue = Queue()
print('Is queue empty? A: '+ str(queue.is_empty()))
queue.queue(1)
queue.queue(2)
queue.queue(3)
queue.queue(4)
queue.queue(5)

print('Queue lenght: ' + str(len(queue)))
print('Last queue element: ' + str(queue.peek()))
print('Is queue empty? A: '+ str(queue.is_empty()))

print('Dequeued element: ' + str(queue.dequeue()))
print('First queue element: ' + str(queue))
print('Last queue element: ' + str(queue.peek()))
print('Queue lenght: ' + str(len(queue)))
```
